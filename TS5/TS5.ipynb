## Estimación espectral

Graficaremos una senoidal, pero le agregaremos una desintonia respecto a $\Delta_f$. Estos seran:

$$k_0 = \{ N/4, N/4 + 0.25, N/4 + 0.5 \}$$

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig

def my_senoidal (N, freq_M, amplitud = 1, valor_medio = 0, freq = 1, fase = 0):
    
    ts = 1/freq_M
    
    tt = np.linspace(0, (N-1)*ts, N)
    
    xx = amplitud * np.sin(2*np.pi*(freq)*tt + fase) + valor_medio
    
    return(tt,xx)

fs = 1000 # frecuencia de muestreo
N = 1000   # cantidad de muestras
DC = 0
Amplitud = np.sqrt(2)

k0 = N/4
freq = k0*fs/N
freq_1 = (k0+0.25)*fs/N
freq_2 = (k0+0.5)*fs/N
#freq_3 = (k0+0.75)*fs/N

zero_padd = 0

# genero mis senoidales
[t, Sr] = my_senoidal(N,fs,Amplitud, DC,freq)
[t, Sr_1] = my_senoidal(N,fs,Amplitud, DC,freq_1)
[t, Sr_2] = my_senoidal(N,fs,Amplitud, DC,freq_2)
#[t, Sr_3] = my_senoidal(N,fs,Amplitud, DC,freq_3)

# realizo la fft de las senoidales
Sr_fft = np.fft.fft(Sr)/Sr.shape[0]
Sr1_fft = np.fft.fft(Sr_1)/Sr_1.shape[0]
Sr2_fft = np.fft.fft(Sr_2)/Sr_2.shape[0]
#Sr3_fft = np.fft.fft(Sr_3)/Sr_3.shape[0]

# armo el vector para la tf
ff_p = np.arange(0, fs, fs/((zero_padd+1)*N))

bfrec = ff_p <= fs/2

#calculo su area
Area = np.sum(2*np.abs(Sr_fft[bfrec])**2)
Area_1 = np.sum(2*np.abs(Sr1_fft[bfrec])**2)
Area_2 = np.sum(2*np.abs(Sr2_fft[bfrec])**2)
#Area_3 = np.sum(2*np.abs(Sr3_fft[bfrec])**2)

# ploteo su desnsidad espectral
plt.clf()

fig, ax = plt.subplots(nrows=1 , ncols=1)
fig.set_size_inches(16,12)
ax.set_title('Densidad Espectral de Potencia')
ax.set_ylabel('Amplitud')
ax.set_xlabel('frecuencia [Hz]')
ax.plot( ff_p[bfrec], (2*np.abs(Sr_fft[bfrec])**2),label =  "$f_0$ = {:3.2f} - $K_0$ ".format(freq,Area))
ax.plot( ff_p[bfrec], (2*np.abs(Sr1_fft[bfrec])**2),label = "$f_0$ = {:3.2f} - $K_0$ + 0.25".format(freq_1))
ax.plot( ff_p[bfrec], (2*np.abs(Sr2_fft[bfrec])**2),label = "$f_0$ = {:3.2f} - $K_0$ + 0.5".format(freq_2))
ax.grid()
ax.set_xlim([245,255])
ax.legend()
```

```python
print("la señal Ko tiene un area de: %.2f"%Area)
print("la señal Ko + 0.25 tiene un area de: %.2f"%Area_1)
print("la señal Ko + 0.50 tiene un area de: %.2f"%Area_2)
```

Podemos ver que estas desintonias agregadas a nuestra señal provocan un cambio en la visualizacion de sus densidades espectrales. Esperabamos ver 3 "deltas", pero las 3 señales tienen graficos distintos pero todas poseen el misma area.
Este fenómeno se debe a que las frecuencias de las señales resultantes de la desintonia agregada no son multiplos de la densidad espectral. Nos quedamos sin **resolución** necesaria para poder representarlas. Por eso los graficos son diferentes. 
Tambien se puede notar que alcanzamos una amplitud unitaria para K0 sin embargo, para las otras señales, la amplitud se redujo y se ensancharon. Podria decirse que se **"desparramaron"** 

## Zero padding

Agregaremos ficticiamente ceros a nuestra señal original en el tiempo, a fin de poder aumetar $\Delta_f$. Usaremos un vector de ceros del tamaño 9*N

```python
zero_padd = 9

# realizo el zero padding
Sr = np.append(Sr,np.zeros(zero_padd*N))
Sr_1 = np.append(Sr_1,np.zeros(zero_padd*N))
Sr_2 = np.append(Sr_2,np.zeros(zero_padd*N))

Sr_fft = np.fft.fft(Sr)/Sr.shape[0]
Sr1_fft = np.fft.fft(Sr_1)/Sr_1.shape[0]
Sr2_fft = np.fft.fft(Sr_2)/Sr_2.shape[0]

# armo el vector para la tf
ff_p = np.arange(0, fs, fs/((zero_padd+1)*N))

bfrec = ff_p <= fs/2

#calculo su area
Area = np.sum(2*np.abs(Sr_fft[bfrec])**2)
Area_1 = np.sum(2*np.abs(Sr1_fft[bfrec])**2)
Area_2 = np.sum(2*np.abs(Sr2_fft[bfrec])**2)

# ploteo su desnsidad espectral
plt.clf()

fig, ax = plt.subplots(nrows=1 , ncols=1)
fig.set_size_inches(16,12)
ax.set_title('Densidad Espectral de Potencia')
ax.set_ylabel('Amplitud')
ax.set_xlabel('frecuencia [Hz]')
ax.plot( ff_p[bfrec], (2*np.abs(Sr_fft[bfrec])**2),label =  "$f_0$ = {:3.2f} - $K_0$ ".format(freq,Area))
ax.plot( ff_p[bfrec], (2*np.abs(Sr1_fft[bfrec])**2),label = "$f_0$ = {:3.2f} - $K_0$ + 0.25".format(freq_1))
ax.plot( ff_p[bfrec], (2*np.abs(Sr2_fft[bfrec])**2),label = "$f_0$ = {:3.2f} - $K_0$ + 0.5".format(freq_2))
ax.grid()
ax.set_xlim([245,255])
ax.legend()
```

Ahora podemos llegar a apreciar donde se encontraria la frecuencia de nuestras senoidales. Ganamos resolución. 
Otra conclusion importante que podemos notar es que dejamos en evidencia la ventana que utiliza la fft. En este caso, una **rectangular** que su fft es una sinc(se utiliza una aproximación, Kernel de Dirichlet)
