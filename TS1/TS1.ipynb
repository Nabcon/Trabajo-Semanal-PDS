### Actividad
Se pedia realizar un generador de se単ales senoidales parametrizables.

```python
import numpy as np
import matplotlib.pyplot as plt


#N = 1000
#fs = 1000

def my_senoidal (N, freq_M, amplitud = 1, valor_medio = 0, freq = 1, fase = 0):
    
    ts = 1/freq_M
    
    tt = np.linspace(0, (N-1)*ts, N)
    
    xx = amplitud * np.sin(2*np.pi*(freq)*tt + fase) + valor_medio
    
    return(tt,xx)

[t, y] = my_senoidal(N=100,freq_M = 100)

fig, ax = plt.subplots(nrows=1 , ncols=1)
fig.set_size_inches(16,12)

ax.plot(t,y,linewidth = 3)
ax.set_ylabel('Amplitud [V]')
ax.set_xlabel('Tiempo [Seg]')
ax.grid()
```

La se単al senoidal se encuentra normalizada. Podemos ver que la forma de salida depende ampliamente de los parametros que definamos. Ahora ya que tenemos libertad absoluta en cuanto a sus parametros, podemos ver como cambia la se単al de salida alterando alguno de ellos.

#### Offset de continua

```python
[t_off, y_off] = my_senoidal(N=100,freq_M = 100,valor_medio=1)

fig, ax = plt.subplots(nrows=1 , ncols=1)
fig.set_size_inches(16,12)

ax.plot(t,y,linewidth = 3)
ax.plot(t_off,y_off, color = 'red',linewidth = 3)
ax.set_ylabel('Amplitud [V]')
ax.set_xlabel('Tiempo [Seg]')
ax.grid()
```

#### Cambios de fase

```python
[t_45, y_45] = my_senoidal(N=100,freq_M = 100,fase = -np.pi/4)
[t_90, y_90] = my_senoidal(N=100,freq_M = 100,fase = np.pi/2)

fig, ax = plt.subplots(nrows=1 , ncols=1)
fig.set_size_inches(16,12)

ax.plot(t,y,label='$\phi_0 = 0属$', linewidth = 3)
ax.plot(t_45,y_45, color = 'red',label='$\phi_1 = -45属$',linewidth = 3)
ax.plot(t_90,y_90, color = 'purple',label='$\phi_2 = 90属$',linewidth = 3)
ax.set_ylabel('Amplitud [V]')
ax.set_xlabel('Tiempo [Seg]')
ax.grid()
ax.legend()
```

#### Distintas frecuencias

```python
[t_3, y_3] = my_senoidal(N=100,freq_M = 100,freq = 3)
[t_6, y_6] = my_senoidal(N=100,freq_M = 100,freq = 6)

fig, ax = plt.subplots(nrows=1 , ncols=1)
fig.set_size_inches(16,12)

ax.plot(t,y,label='$f = f_0$', linewidth = 3)
ax.plot(t_3,y_3, color = 'red',label='$f_2 = 3*f_0$',linewidth = 3)
ax.plot(t_6,y_6, color = 'purple',label='$f_3 = 6*f_0$',linewidth = 3)
ax.set_ylabel('Amplitud [V]')
ax.set_xlabel('Tiempo [Seg]')
ax.grid()
ax.legend()
```

Podemos notar que las se単ales de mas alta frecuencia, con la misma configuracion de N y Fs que la se単al original, no tienen una curvas suaves. Esto se debe a que no al aumentar la frecuencia, tengo menos puntos para poder representar a esas se単ales mas rapidas. Lo que puedo hacer es **aumentar la frecuencia de muestreo y el N**.

```python
[t_3, y_3] = my_senoidal(N=300,freq_M = 300,freq = 3)
[t_6, y_6] = my_senoidal(N=300,freq_M = 300,freq = 6)

fig, ax = plt.subplots(nrows=1 , ncols=1)
fig.set_size_inches(16,12)

ax.plot(t,y,label='$f = f_0$', linewidth = 3)
ax.plot(t_3,y_3, color = 'red',label='$f_2 = 3*f_0$',linewidth = 3)
ax.plot(t_6,y_6, color = 'purple',label='$f_3 = 6*f_0$',linewidth = 3)
ax.set_ylabel('Amplitud [V]')
ax.set_xlabel('Tiempo [Seg]')
ax.grid()
ax.legend()
```

Podemos notar que las curvas se suavizaron.

#### Ripple
Tambien podemos jugar y simular el ripple de 50Hz que existe en una se単al continua con ruido

```python
[t_rp, y_rp] = my_senoidal(N=1000,freq_M = 1000,freq = 50, amplitud = 0.01, valor_medio=5)

fig, ax = plt.subplots(nrows=1 , ncols=1)
fig.set_size_inches(16,12)

ax.plot(t_rp,y_rp,label='$f = 50Hz$', linewidth = 3)
ax.set_ylabel('Amplitud [V]]')
ax.set_xlabel('Tiempo [Seg]')
ax.set_ylim([0, 6])
ax.grid()
ax.legend()
```
